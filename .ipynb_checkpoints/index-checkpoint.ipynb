Las datas frames elegidas corresponden a las siguientes páginas web: 
1. Population by country: 
2. Green House Emissions by country: 
3. Freshwater withdrawals by country: 

```python
#Links to databases in folder "data" in repository "Evalucion1"
population_Link="https://www.cia.gov/the-world-factbook/field/population/country-comparison/"
h2O_withdrawl_Link="https://github.com/SpatialDataScience23-2/Evaluacion1/raw/main/data/freshwater_withdrawals.csv"
pm2_5_Link="https://github.com/SpatialDataScience23-2/Evaluacion1/raw/main/data/pm2_5.csv"

import pandas as pd
population=pd.read_html(population_Link, header=0, flavor= 'bs4')
h2O_withdrawl=pd.read_csv(h2O_withdrawl_Link)
pm2_5=pd.read_csv(pm2_5_Link)

population_new=population[0]
```

### df1: h2O_withdrawl
1. In this data base we have the information of the total annual freswater withdrawls by each country from 1975-2019
2. That's alot of data, so I decided to restrict it to just the variable values of year 2018
3. We recall the reduced data base with the name "h2O"

```python
h2O_new=h2O_withdrawl.copy()
h2O_new.Year.value_counts()
```

```python
#we restrict the Year value to only 2018
h2O_new_1=h2O_new[h2O_new["Year"]==2018]
h2O_new_1.Year.value_counts()
#we eliminate the column "Year"
h2O=h2O_new_1.drop(columns="Year")
h2O
```

### df2: PM2.5
1. In this data base we have the information of the PM2.5 by each country

```python
pm2_5.Year.value_counts()
```

```python
pm2_5_new=pm2_5.copy()
pm2_5_new_1=pm2_5_new[pm2_5_new["Year"]==2018]
pm2_5_new_1.YEA.value_counts()
```

```python
pm25=pm2_5_new_1.copy()
byeColumns=['Year','YEA']
pm25.drop(columns=byeColumns,inplace=True)
pm25
```

### Cleaning & formatting
Now we will procede with the cleaning and formatting (if needed)

The variables we want to keep are: 
1. Value: this corresponds to the Mean population exposure to PM2.5. Its units are Micrograms per cubic metre. Note that initially we restricted the original "PM2.5" data frame to only the values from 2018. 
2. Population: It is the number of people by country
3. Annual freshwater withdrawals, total (billion cubic meters): Note that initially we restricted the original "Annual freshwater withdrawls" to the values from year 2018. 

Therefore: 
For column "Value", we will rename it to "PM2.5_exposure_2018"
For column "Annual freshwater withdrawals, total (billion cubic meters)", we will rename it to "Annual_H2O_withdrawl_2018"

```python
# We get rid off of any white spaces between the words of the columns' titles
final_df.columns=final_df.columns.str.replace(" ","_")
final_df
```

```python
# We get rid off of the variables that are not of our interest
delete_columns=["COU","SMALL_SUBNATIONAL_REGION","Small_subnational_region","LARGE_SUBNATIONAL_REGION","Large_subnational_region",
               "Reference_Period_Code","Reference_Period","Flag_Codes","Flags","Rank","Code"]
final_df.drop(columns=delete_columns,inplace=True)
final_df
```

```python
# We keep deleting the unnecesary columns
bye_columns=["VAR","Variable","Unit_Code","Unit","PowerCode_Code","PowerCode"]
final_df.drop(columns=bye_columns,inplace=True)
final_df
```

```python
# We rename certain column titles
final_df.rename(columns={'Value':'PM2.5_exposure_2018', 'Date_of_Information':'Year_of_population_value',
                        'Annual_freshwater_withdrawals,_total_(billion_cubic_meters)':'Annual_H2O_withdrawl_2018'},inplace=True)
final_df
```

```python
final_df.info()
```

```python
# In column "Year_of_population_value", we need to get rid off est. and white spaces
# Note that python reads the values as objects instead of numbers
final_df['Year_of_population_value']=final_df.Year_of_population_value.str.extract(pat=r'(^\w+|.$)', expand=True)
final_df
```

```python
#Since all values correspond to year 2023, we can delete that column
final_df=final_df.drop('Year_of_population_value', axis=1)
```

```python
# We rename the population column name
final_df.rename(columns={'Population':'Population_in_2023','PM2.5_exposure_2018':'PM2_5_exposure_2018'},inplace=True)
final_df
```

```python
# Lastly we corroborate that the columns correspond to the variable type (string or object, float and int)
final_df.info()
```
